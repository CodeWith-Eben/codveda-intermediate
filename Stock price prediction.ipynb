```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('Stock Prices Data Set.csv',index_col=0)
```

```python
df.head()
```

```python
df = df.reset_index()
```

```python
print(df.dtypes)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
print(df.isnull().sum())
```

```python
df['open'] = df['open'].ffill()
df['high'] = df['high'].ffill()
df['low']  = df['low'].ffill()
```

```python
print(df.isnull().sum())
```

```python
df['symbol'] = df['symbol'].str.strip().str.upper()
```

```python
invalid_rows = df[df['high'] < df['low']]
print("Invalid high/low rows:", len(invalid_rows))
```

```python
df = df[df['high'] >= df['low']]
```

```python
zero_check = (df[['open', 'high', 'low', 'close']] == 0).sum()
print("Zero values:\n", zero_check)
```

```python
negative_check = (df[['open', 'high', 'low', 'close', 'volume']] < 0).sum()
print("Negative values:\n", negative_check)
```

```python
x = df[['low', 'high', 'close', 'volume']]
y = df['open']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, 
    test_size=0.2,     
    random_state=42     
)

print("Training size:", x_train.shape)
print("Testing size:", x_test.shape)
```

```python

model = LinearRegression()
model.fit(x_train, y_train)

print("Model trained successfully!")
```

```python
y_pred = model.predict(x_test)
print("Predictions:", y_pred[:5])
print("Actual:", y_test[:5].values)
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MSE:  {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²:   {r2:.2f}")
```
